```
# Authors: Umesh Nimmathi - 1649330, Sachiketha Pasupula Reddy - 1629720
# Program to find whether a person is diabetic or not
# Reference : https://www.hindawi.com/journals/jdr/2019/4248218/

import ipywidgets as widgets
node=0
Questions = ['<b>Is your age greater than 49?</b>',
             '<b>Is your BMI greater than or equal to 25?</b>',
             '<b>Does any of your family has diabetes?<b>',
             '<b>Is your age greater than 34<b>?',
             '<b>Do you have work-related stress?<b>',
              '<b>Do you work out or do physical activity?<b>',
              '<b>Do you have hypertension?<b>',
              '<b>Do you have family history of diabetes?<b>',
             '<b>Do you have history of cardiovascular disease or stroke?<b>',
             '<b>Do you take salty food ?<b>',
              '<b>Are you Male ?<b>']  #Questions to post 

label = widgets.HTML(
    value="<b>Is your age greater than 49?</b>",
    placeholder='',
    description='',
)                            # Label Widget to display questions
button1 = widgets.Button(
    description='Yes',
    disabled=False,
    button_style='', 
    tooltip='Click me',
    icon='check' # Button for Yes
)
test_again= widgets.Button(
    description='Test Again',
    disabled=False,
    button_style='', 
    tooltip='Click me',
    icon='check' # Button for Test again
)
button2= widgets.Button(
    description='No',
    disabled=False,
    button_style='', 
    tooltip='Click me',
    icon='check' # Button for No
)

display(label,button1,button2,test_again)

def clear_widgets():          # Disables buttons 
  global button1,button2,test_again
  button1.disabled=True
  button2.disabled=True
    
def start_again(b):
  global node
  label.value=Questions[0]
  node=0
  button1.disabled=False
  button2.disabled=False
  
  
def on_button_clicked(b):   
 global node
 if(node==0 and b.description=='No'):
      label.value=Questions[1]
      node=1  
 elif(node==1 and b.description=='Yes'):
       label.value=Questions[2]  # if Age<=49 and BMI >= 25
       node=2
 elif(node==1 and b.description=='No'):
      label.value= "Non-Diabetes"# if Age <=49 and BMI<25 
      clear_widgets()
      node=0  
 elif(node==2 and b.description=='Yes'):
      label.value=Questions[3]
      node=3  
 elif(node==3 and b.description=='Yes'):
      label.value='Diabetes'
      node=0                       #if Family_History is Yes and age>34 
      clear_widgets()
 elif(node==3 and b.description=='No'):
      label.value='Non-Diabetes'
      node=0                       #if Family_History is Yes and age<=34 
 elif(node==2 and b.description=='No'):
      label.value=Questions[3]      #if Family_History is 'No' go to node Age
      node=4                       
 elif(node==4 and b.description=='No'):
      label.value='Non-Diabetes'
      node=0                       #if Family_History is No and age<=34
      clear_widgets()
 elif(node==4 and b.description=='Yes'):
      label.value=Questions[4]     #if age>=34 and work_stress is 'Yes' , go to node workstress
      node=5 
 elif(node==5 and b.description=='Yes'):
      label.value=Questions[5]     #if age>=34 and work_stress is 'Yes' go to activity
      node=6 
 elif(node==6 and b.description=='Yes'):
      label.value='Non-diabetes'     #if  Activity is 'Yes' then Non-diabetes
      node=0
      clear_widgets()
 elif(node==6 and b.description=='No'):
      label.value='Diabetes'     #if Activity is 'No' then Diabetes
      node=0 
      clear_widgets()
 elif(node==5 and b.description=='No'):
      label.value=Questions[6]     #if age>=34 and work_stress is 'Yes' go to Hypertension node
      node=7 

 elif(node==7 and b.description=='Yes'):
      label.value=Questions[5]    
      node=9   # if hypertension is 'Yes' then go to Activity node
 elif(node==7 and b.description=='No'):
      label.value=Questions[8]    
      node=8   # if hypertension is 'No' then go to History node
 elif(node==8 and b.description=='No'):
      label.value='Non-diabetes'     #if  History is 'No' then Non-diabetes
      node=0
      clear_widgets()
 elif(node==8 and b.description=='Yes'):
      label.value='Diabetes'     #if History is 'Yes' then Diabetes
      node=0 
 elif(node==9 and b.description=='Yes'):
      label.value='Non-diabetes'     #if  Activity is 'Yes' then Non-diabetes
      node=0
      clear_widgets()
 elif(node==9 and b.description=='No'):
      label.value='Diabetes'     #if Activity is 'No' then Diabetes
      node=0 
      clear_widgets()

 elif(node==0 and b.description=='Yes'):
      label.value=Questions[1]   #if Age>49 go to BMI node 
      node=10 
 elif(node==10 and b.description=='Yes'):  #if age >49 and BMI >=25 then go to work related stress node 
      label.value=Questions[4]
      node=11
 elif(node==10 and b.description=='No'):  #if age >49 and BMI < 25 then go to activity node
      label.value=Questions[5]
      node=12

 elif(node==11 and b.description=='No'): 
      label.value='Non-Diabetes'  # if Work sress in 'No' then Non-Diabetes
      node=0
 elif(node==11 and b.description=='Yes'):
      label.value='Diabetes'      # if Work sress in 'Yes' then Diabetes
      node=0
      clear_widgets()
 elif(node==12 and b.description=='No'):  # if Activity is no then go to node 13
      label.value=Questions[9]
      node=13
 elif(node==12 and b.description=='Yes'):
      label.value='Non-Diabetes'      #if activity is 'Yes' then Non-Diabetes
      node=0
      clear_widgets()
 elif(node==13 and b.description=='Yes'):
      label.value=Questions[4]         # if 'Salty food' is 'yes' then go to node 14
      node=14
 elif(node==13 and b.description=='No'):
      label.value='Non-Diabetes'  # if 'Salty food' is 'No'then 'Non-Diabetes'
      node=0
      clear_widgets()

 elif(node==14 and b.description=='No'):
      label.value=Questions[6]         # if 'Work related stress' is 'No' then go to node 15
      node=15
 elif(node==14 and b.description=='Yes'):
      label.value= 'Diabetes'       # if 'Work related stress' is 'No' then go to node 15
      node=0
      clear_widgets()

 elif(node==15 and b.description=='No'):
      label.value=Questions[10]         # if 'Hyper tension' is 'No' then go to node 16
      node=16  
 elif(node==15 and b.description=='Yes'):
      label.value= 'Diabetes'         # if 'Hyper tension' is 'Yes' then Diabetes.
      node=0
      clear_widgets()

 elif(node==16 and b.description=='No'):
      label.value=Questions[8]         # if you are not men then go to 'histoty of cardiovascular disease node'
      node=17
 elif(node==16 and b.description=='Yes'):
      label.value='Diabetes'         # if you are 'Men' then 'Diabetes'
      node=0
      clear_widgets()
 elif(node==17 and b.description=='No'):
      label.value='Non-Diabetes'    # if you dont have any history of cardio vascular diesease then Non-Diabetes
      node=0
      clear_widgets()
 elif(node==17 and b.description=='Yes'):
      label.value='Diabetes'    # if you have any history of cardio vascular diesease then 'Diabetes'
      node=0 
      clear_widgets()


 print(node,':')
button1.on_click(on_button_clicked)
button2.on_click(on_button_clicked)
test_again.on_click(start_again)
```
